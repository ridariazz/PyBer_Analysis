```python
# Add inline on the first line 
%matplotlib inline

# Import the following dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Plot the months along the x-axis and the fare on the y-axis using plot() fxn
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Adjust the x-ticks to show all the months 

# Set x-axis and tick locations 
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# two ways to create a bar chart with the same data 
# chain bar() function to the plot() function 

pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
#  other approach is to add the kind parameter to the plot() function. 
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```
